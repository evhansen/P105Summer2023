```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

```

### Set some initial conditions

```python
#change me!
m = 3.5
X = 15.0
P = 10.0
```

```python
#do not alter these conditions (per the book)
t0 = 0 
t = np.arange(0,10,1)
x0a = 0 
p0a = 0 

x0b = X 
p0b = 0 

x0c = X
p0c = P

x0d = 0 
p0d = P

def x_func(t, p0, x0):
    return 0.5*10*t**2 + p0*t/m + x0

def p_func(t, p0):
    return m*10*t + p0
```

### Plot the Position

```python
plt.scatter(t, x_func(t, p0a, x0a),label='A')
plt.scatter(t, x_func(t, p0b, x0b),label='B')
plt.scatter(t, x_func(t, p0c, x0c),label='C')
plt.scatter(t, x_func(t, p0d, x0d),label='D')


plt.xlabel('position')
plt.ylabel('time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
```

### Plot the momentum

```python
plt.scatter(t, p_func(t, p0a),label='A')
plt.scatter(t, p_func(t, p0b),label='B')
plt.scatter(t, p_func(t, p0c),label='C')
plt.scatter(t, p_func(t, p0d),label='D')


plt.xlabel('momentum')
plt.ylabel('time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
```

### Visualizing Phase Space

```python
plt.scatter(x_func(t, p0a, x0a), p_func(t, p0a),label='A')
plt.scatter(x_func(t, p0b, x0b), p_func(t, p0b),label='B')
plt.scatter(x_func(t, p0c, x0c), p_func(t, p0c),label='C')
plt.scatter(x_func(t, p0d, x0d), p_func(t, p0d),label='D')

plt.plot(x_func(t, p0a, x0a), p_func(t, p0a),label='A')
plt.plot(x_func(t, p0b, x0b), p_func(t, p0b),label='B')
plt.plot(x_func(t, p0c, x0c), p_func(t, p0c),label='C')
plt.plot(x_func(t, p0d, x0d), p_func(t, p0d),label='D')

plt.xlabel('position')
plt.ylabel('momentum')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.legend()
```

Remember to have it grainy enough that the lines are continuous --- if you don't have fine enough binning it will _look_ like the lines intersect

### Mapping out the areas in phase space

```python
def polygon_area(x,y):
    correction = x[-1] * y[0] - y[-1]* x[0]
    main_area = np.dot(x[:-1], y[1:]) - np.dot(y[:-1], x[1:])
    return 0.5*np.abs(main_area + correction)
```

```python
fig, ax = plt.subplots()

x = [x_func(t[0], p0a, x0a),x_func(t[0], p0b, x0b),x_func(t[0], p0c, x0c),x_func(t[0], p0d, x0d)]
y = [p_func(t[0], p0a),p_func(t[0], p0b),p_func(t[0], p0c),p_func(t[0], p0d)]
ax.add_patch(patches.Polygon(xy=list(zip(x,y)), edgecolor ='k',facecolor='whitesmoke'))
plt.text(x_func(t[0], p0a, x0a)+12, p_func(t[0], p0a)+5, 'Area = %0.2f'%polygon_area(x,y))

x = [x_func(t[1], p0a, x0a),x_func(t[1], p0b, x0b),x_func(t[1], p0c, x0c),x_func(t[1], p0d, x0d)]
y = [p_func(t[1], p0a),p_func(t[1], p0b),p_func(t[1], p0c),p_func(t[1], p0d)]
ax.add_patch(patches.Polygon(xy=list(zip(x,y)), edgecolor ='k',facecolor='whitesmoke'))
plt.text(x_func(t[1], p0a, x0a)+15, p_func(t[1], p0a)+5, 'Area = %0.2f'%polygon_area(x,y))


x = [x_func(t[2], p0a, x0a),x_func(t[2], p0b, x0b),x_func(t[2], p0c, x0c),x_func(t[2], p0d, x0d)]
y = [p_func(t[2], p0a),p_func(t[2], p0b),p_func(t[2], p0c),p_func(t[2], p0d)]
ax.add_patch(patches.Polygon(xy=list(zip(x,y)), edgecolor ='k',facecolor='whitesmoke'))
plt.text(x_func(t[2], p0a, x0a)+18, p_func(t[2], p0a)+5, 'Area = %0.2f'%polygon_area(x,y))


x = [x_func(t[3], p0a, x0a),x_func(t[3], p0b, x0b),x_func(t[3], p0c, x0c),x_func(t[3], p0d, x0d)]
y = [p_func(t[3], p0a),p_func(t[3], p0b),p_func(t[3], p0c),p_func(t[3], p0d)]
ax.add_patch(patches.Polygon(xy=list(zip(x,y)), edgecolor ='k',facecolor='whitesmoke'))
plt.text(x_func(t[3], p0a, x0a)+21, p_func(t[3], p0a)+5, 'Area = %0.2f'%polygon_area(x,y))


plt.scatter(x_func(t, p0a, x0a), p_func(t, p0a))
plt.scatter(x_func(t, p0b, x0b), p_func(t, p0b))
plt.scatter(x_func(t, p0c, x0c), p_func(t, p0c))
plt.scatter(x_func(t, p0d, x0d), p_func(t, p0d))
plt.plot(x_func(np.arange(0,10,.1), p0a, x0a), p_func(np.arange(0,10,.1), p0a),label='A')
plt.plot(x_func(np.arange(0,10,.1), p0b, x0b), p_func(np.arange(0,10,.1), p0b),label='B')
plt.plot(x_func(np.arange(0,10,.1), p0c, x0c), p_func(np.arange(0,10,.1), p0c),label='C')
plt.plot(x_func(np.arange(0,10,.1), p0d, x0d), p_func(np.arange(0,10,.1), p0d),label='D')


plt.xlabel('position')
plt.ylabel('momentum')
plt.legend(loc='upper left')
plt.title('X=%0.2f, P=%0.2f, m=%0.2f'%(X,P,m))

plt.xlim(-1,x_func(t[4], p0d, x0d))
plt.ylim(-1,p_func(t[3], p0d)+5)

```
